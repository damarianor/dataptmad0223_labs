#### Plot the correlation matrix using seaborn.

```python
corr = titanic.corr(method='pearson')
```

```python
sns.set(style='dark')                           
ma =np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(0, 10, as_cmap=True)
sns.heatmap(corr,             
            mask=ma,    
            cmap=cmap,
            vmax=1,
            vmin=-1,
            center=0,
            square=True,
            linewidth=0.5,
            cbar_kws={'shrink': 0.5},
            annot=True
           );
```
