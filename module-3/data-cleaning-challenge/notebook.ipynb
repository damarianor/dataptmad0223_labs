```python
data['class'] = data['class'].astype('str')
```

```python
data.describe()
```

```python
cols = ['class']

data_list = []
for col in cols:
    data_unique = data[col].unique()
    data_num = len(data_unique)
    data_dict = {
        "categorical_variable": col,
        "number_of_possible_values": data_num,
        "values": data_unique.tolist()
    }
    data_list.append(data_dict)

categories = pd.DataFrame(data_list).sort_values(
    by="number_of_possible_values", ascending=False
).reset_index(drop=True)

categories
```

```python
data.isna().sum()
```

```python
def missing_percentage(df):
    percent_missing = df.isnull().sum() * 100 / len(df)
    missing_values_df = pd.DataFrame({'column_name': df.columns,'percent_missing': percent_missing})
    return missing_values_df
```

```python
missing_percentage(data)
```

```python
no_nan_col = ['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm']
data_no_nan_col = data[no_nan_col]
data_no_nan_col
```

```python
missing_percentage(data_no_nan_col)
```

```python
data_no_nan_rows = data.dropna()
data_no_nan_rows
```

```python
missing_percentage(data_no_nan_rows)
```
